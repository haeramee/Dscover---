```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import mean_squared_error
```

```python
data_org = pd.read_csv("/content/drive/MyDrive/2022-1/2022-1 DScover/01. 태양광 프로젝트/태양광 프로젝트!/merged_final.csv")
data = data_org.copy()
data.head()
```

```python
data['day'] = data['date'].str.split('-').str[2]
data['month'] = data['date'].str.split('-').str[1]
data['year'] = data['date'].str.split('-').str[0]
```

```python
data
```

```python
def area_mean(item) : 
  df = data[data.area == item]
  df = df.drop(['area', 'date', 'generation','day','year'], axis=1)
  df = df.groupby(['month'])['avgTa','avgWs','avgRhm','ssDur','sumSsHr','sumGsr','avgTca','PM10','PM25'].mean()
  print("{}".format(item))
  return df
```

```python
area_mean('세종시')
# 위 함수에 원하는 지역 이름을 입력하면 해당 지역의 월별 5년치 평균 데이터를 data frame으로 봔환
# ['강원도', '경기도','경상남도','경상북도','광주시','대구시','대전시','부산시','서울시','세종시','울산시','인천시','전라남도','전라북도','제주도','충청남도','충청북도']
```
